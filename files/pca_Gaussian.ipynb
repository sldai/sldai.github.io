```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import numpy.random as rnd
from matplotlib.patches import Ellipse

NUM = 2

ells = [Ellipse(xy=rnd.rand(2)*10, width=rnd.rand(), height=rnd.rand(), angle=rnd.rand()*360, fill=False, ls='--')
        for i in range(NUM)]

fig = plt.figure(0)
ax = fig.add_subplot(111, aspect='equal')
for e in ells:
    ax.add_artist(e)
    e.set_clip_box(ax.bbox)
    # e.set_alpha(rnd.rand())
    e.set_facecolor(rnd.rand(3))
    e.set_edgecolor(rnd.rand(3))

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

plt.show()
```

```python
mean = [0.0,0.0]
eig_vec = np.array([[1,1],
                [1,-1]],dtype=float)
eig_val = np.diag([2,1])
cov = eig_vec @ eig_val @ eig_vec.T
x, y = np.random.multivariate_normal(mean, cov, 200).T

fig = plt.figure(0)
ax = fig.add_subplot(111, aspect='equal')
ax.plot(x, y, '+', color='red')
e = Ellipse(xy=mean, width=eig_val[0,0]*4, height=eig_val[1,1]*4, angle=np.arctan2(eig_vec[1,0],eig_vec[0,0])*180/np.pi, fill=False, ls='--')
ax.add_artist(e)
e.set_clip_box(ax.bbox)
e.set_edgecolor('blue')

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

plt.savefig('ellipse.png')
```
